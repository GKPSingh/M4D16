```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
x
```

```python
y =  x**2
y
```

```python
# Functional
plt.plot(x,y);
plt.xlabel('X-axis'), plt.ylabel('Y-axis'), plt.title('Simple Plot');
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x, 'b');
```

```python
x = np.linspace(0,10,15)
y =  x**2

fig, ax = plt.subplots(figsize = (10,8))
ax.plot(x,y)
ax.set(Title = 'Simple Plot', ylabel = 'Y-axis', Xlabel = 'X-axis');
ax.set(xlim=(0), ylim=(0));
```

```python
fig,(ax0, ax1)= plt.subplots(nrows = 2, ncols =1)
ax0.plot(x,y, 'r *')
ax1.plot(y,x, 'k--')
ax0.set(xlim=(0), ylim=(0));
ax1.set(xlim=(0), ylim=(0));
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.18, 0.55, 0.4, 0.3])
axes1.plot(x,y,'r*')
axes2.plot(y,x, 'k--')
axes1.set(Title ='XY plot', xlabel = 'X-Values', ylabel = 'Y-Values', xlim=0, ylim = 0);
axes2.set(Title ='YX plot', xlabel = 'Y-Values', ylabel = 'X-Values', xlim=0, ylim = 0);
fig.suptitle('Yahoo Plot!!!', fontsize =16, fontweight = 'bold');
```
